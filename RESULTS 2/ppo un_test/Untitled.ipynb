```python
import pandas as pd
import numpy as np
```

```python
def delt(f1,f2):
    #f='./ppo2 nosafe_test_result/Rain0_randomlevel5_ppo2 nosafeflooding_vs_t.csv'
    data1=pd.read_csv(f1).values
    data2=pd.read_csv(f2).values
    data=np.concatenate((data1,data2),axis=1)
    #print(np.max(data[-1,:]))
    #print(np.min(data[-1,:]))
    print('        ',np.max(data[-1,:])-np.min(data[-1,:]))
```

```python
for rainid in range(10):
    print(rainid)
    for rd in [5,10,15,20]:
        print('   ',rd)
        f1='./ppo2 nosafe_test_result/Rain'+str(rainid)+'_randomlevel'+str(rd)+'_ppo2 nosafeflooding_vs_t.csv'
        f2='./ppo2 nosafe_test_result - 1/Rain'+str(rainid)+'_randomlevel'+str(rd)+'_ppo2 nosafeflooding_vs_t.csv'
        delt(f1,f2)
```
