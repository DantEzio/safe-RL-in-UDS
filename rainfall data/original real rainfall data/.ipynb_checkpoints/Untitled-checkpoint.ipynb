```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#读取40和80的降雨,40:[3,5,6,7,9],80:[1,4,5,7,9]
rains={}
for rg in ['RG1','RG2','RG3','RG4']:
    rains[rg]=np.loadtxt('./A_real_raindata'+rg+'_80.txt',delimiter=' ')
m,n=rains['RG1'].shape
#
```

```python
#10张图，代表10场降雨，每张图有4条线，代表4个raingage

font1 = {'family' : 'Times New Roman',
         'weight' : 'normal',
         'size'   : 50,}

font2 = {'family' : 'Times New Roman',
         'weight' : 'normal',
         'size'   : 50,}

fig,ax=plt.subplots(10,1,figsize=(80,100))



for i in range(10):
    rmax,rmin=0,0
    for rg in ['RG1','RG2','RG3','RG4']:
        ax[i].plot(range(len(rains[rg][i])),rains[rg][i],label='Rain'+str(i)+' '+rg,linewidth=10)
        if np.max(rains[rg][i])>rmax:
            rmax=np.max(rains[rg][i])
        if np.min(rains[rg][i])<rmin:
            rmin=np.min(rains[rg][i])
    ax[i].legend(prop=font1,loc='right')
    ax[i].set_xticks([])
    ax[i].set_ylabel('Volume (mm)',font2)
    ax[i].set_yticks([rmin,rmax])
    ax[i].set_yticklabels([rmin,rmax],fontsize=40)
plt.xlabel('Time (minute)',font2)
x=['00:00','24:00']
plt.xticks([0,n],x,fontsize=40)

    
plt.savefig('rainfall.png',bbox_inches='tight', dpi=100)
```
