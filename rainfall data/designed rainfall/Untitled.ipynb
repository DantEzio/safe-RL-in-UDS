```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=np.loadtxt('./trainRainFile.txt',delimiter=',')/2.2
```

```python
print(data.shape)
```

```python
font1 = {'family' : 'Times New Roman',
         'weight' : 'normal',
         'size'   : 10,}

font2 = {'family' : 'Times New Roman',
         'weight' : 'normal',
         'size'   : 18,}

plt.figure(figsize=(10,6))
for i in range(10):
    plt.plot(data[i,:],label='Rain'+str(i+1))


plt.legend(prop=font1)
plt.xlabel('Time (minute)',font2)
plt.ylabel('Intensity (mm/min)',font2)
x=['9:00','11:00']
plt.xticks([0,240],x)
plt.savefig('rainfall.png',bbox_inches='tight', dpi=500)
```
